```python
import numpy as np
import pandas as pd
from sklearn.impute import KNNImputer
```

```python
masked = pd.read_csv("assignment/example_data_1_masked.txt", sep = " ")
masked = masked.replace('*', np.NaN)
masked.head()
```

```python
unmasked = pd.read_csv("assignment/example_data_1.txt", sep = " ")
unmasked.shape #(39495, 50)
masked.isna().sum()
```

```python
# Impute
imputeMask = KNNImputer(n_neighbors=2)
imputed = imputeMask.fit_transform(masked)
```

```python
masked = pd.read_csv("assignment/example_data_1_masked.txt", sep = " ", header = None)
masked = masked.replace('*', np.NaN)
masked.head()
```

```python
masked.columns
```

```python
unmasked = pd.read_csv("assignment/example_data_1.txt", sep = " ", header = None)
unmasked.shape #(39495, 50)
masked.isna().sum()
```

```python
# Impute using most common value in column
def fillna(col):
    col.fillna(col.value_counts().index[0], inplace=True)
    return col

maskedImputed=masked.apply(lambda col:fillna(col))
```

```python
maskedImputed.head()
```

```python
unmasked.head()
```

```python
# Another imputer
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy= 'most_frequent')
maskedImputed2=pd.DataFrame(imputer.fit_transform(masked))
```

```python
maskedImputed2.head()
```

```python
maskedImputed.equals(unmasked) #False, there are discrepancies
```

```python
maskedImputed = maskedImputed.astype('int64')
maskedImputed2 = maskedImputed2.astype('int64')
booleanMaskDisc = pd.DataFrame(maskedImputed2 == unmasked)
```

```python
maskedCorrectlyCounts = booleanMaskDisc.apply(pd.Series.value_counts, axis = 0)
```

```python
maskedCorrectlyCounts # over 90% correct when imputed
```

```python
maskedImputed.equals(maskedImputed2)
```
